# Regression Exercise 

California Housing Data

This data set contains information about all the block groups in California from the 1990 Census. In this sample a block group on average includes 1425.5 individuals living in a geographically compact area. 

The task is to aproximate the median house value of each block from the values of the rest of the variables. 

 It has been obtained from the LIACC repository. The original page where the data set can be found is: http://www.liaad.up.pt/~ltorgo/Regression/DataSets.html.
 

The Features:
 
* housingMedianAge: continuous. 
* totalRooms: continuous. 
* totalBedrooms: continuous. 
* population: continuous. 
* households: continuous. 
* medianIncome: continuous. 
* medianHouseValue: continuous. 

## The Data

** Import the cal_housing_clean.csv file with pandas. Separate it into a training (70%) and testing set(30%).**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline 

import tensorflow as tf

from sklearn.model_selection import train_test_split
```

```python
housing = pd.read_csv('cal_housing_clean.csv')
```

```python
housing.shape
```

```python
housing.head(5)
```

```python
housing.describe().T
```

```python
housing.isnull().sum()
```

```python
X = housing[['housingMedianAge', 'totalRooms', 'totalBedrooms', 'population',
       'households', 'medianIncome']]

Y = housing['medianHouseValue']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.3, shuffle=True, random_state=42)
```

```python
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

### Scale the Feature Data

** Use sklearn preprocessing to create a MinMaxScaler for the feature data. Fit this scaler only to the training data. Then use it to transform X_test and X_train. Then use the scaled X_test and X_train along with pd.Dataframe to re-create two dataframes of scaled data.**

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
```

```python
scaler
```

```python
X_train_sc = scaler.fit_transform(X_train)
X_test_sc = scaler.transform(X_test)
```

```python
X_train = pd.DataFrame(X_train_sc, columns=X.columns, index=X_train.index)
X_test = pd.DataFrame(X_test_sc, columns=X.columns, index=X_test.index)
```

```python
X_train.describe().T
```

```python
X_test.describe().T
```

### Create Feature Columns

** Create the necessary tf.feature_column objects for the estimator. They should all be trated as continuous numeric_columns. **

```python
numeric_columns = list(X_train.columns)
print(numeric_columns)
```

```python
numeric_features = [tf.feature_column.numeric_column(key = column) for column in numeric_columns]
```

```python
numeric_features[0]
```

** Create the input function for the estimator object. (play around with batch_size and num_epochs)**

```python
input_func = tf.estimator.inputs.pandas_input_fn(x=X_train, y=y_train, batch_size=32, shuffle=True, num_epochs=1000)
```

** Create the estimator model. Use a DNNRegressor. Play around with the hidden units! **

```python
from tensorflow.estimator import DNNRegressor
```

```python
dnn_model = DNNRegressor(hidden_units=[6, 6, 6], 
    feature_columns = numeric_features, 
    model_dir=None,
    optimizer='Adam',
    dropout=0.2,
    batch_norm=False)
```

##### ** Train the model for ~1,000 steps. (Later come back to this and train it for more and check for improvement) **

```python
dnn_model.train(input_fn=input_func, steps=1000)
```

** Create a prediction input function and then use the .predict method off your estimator model to create a list or predictions on your test data. **

```python
eval_input_func = tf.estimator.inputs.pandas_input_fn(
      x=X_test,
      y=y_test,
      batch_size=32,
      num_epochs=1,
      shuffle=False)
```

```python
dnn_model.evaluate(eval_input_func)
```

```python
pred_input_func = tf.estimator.inputs.pandas_input_fn(x=X_test, batch_size=10, num_epochs=1, shuffle=False)
```

```python
pred = list(dnn_model.predict(pred_input_func))
```

```python
type(pred)
```

```python
pred
```

```python
y_test[:10]
```

```python
prediction = [p['predictions'][0] for p in pred]
```

** Calculate the RMSE. You should be able to get around 100,000 RMSE (remember that this is in the same units as the label.) Do this manually or use [sklearn.metrics](http://scikit-learn.org/stable/modules/generated/sklearn.metrics.mean_squared_error.html) **

```python
prediction[:5]
```

```python
def rmse(predictions, targets):
    return np.sqrt(((predictions - targets) ** 2).mean())
```

```python
rmse(prediction, y_test)
```

```python
from sklearn.metrics import mean_squared_error
```

```python
mean_squared_error(y_test, prediction)**0.5
```
