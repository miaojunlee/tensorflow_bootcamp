** TensorFlow won't be able to understand strings as labels, you'll need to use pandas .apply() method to apply a custom function that converts them to 0s and 1s. This might be hard if you aren't very familiar with pandas, so feel free to take a peek at the solutions for this part.**

** Convert the Label column to 0s and 1s instead of strings.**

```python
df['income_bracket'] = df['income_bracket'].apply(lambda x:1 if x==' >50K' else 0)
```

```python
df['income_bracket'].value_counts()
```

### Perform a Train Test Split on the Data

```python
from sklearn.model_selection import train_test_split
```

```python
X = df.drop(['income_bracket'], axis=1)
Y = df['income_bracket']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,Y,test_size=0.3, shuffle=True, random_state=42)
```

### Create the Feature Columns for tf.esitmator

** Take note of categorical vs continuous values! **

```python
df.columns
```

** Import Tensorflow **

```python
import tensorflow as tf
```

** Create the tf.feature_columns for the categorical values. Use vocabulary lists or just use hash buckets. **

```python
[df[col].nunique() for col in categorical_columns]
```

```python
df['occupation'].nunique()
```

```python
# may consider using different ways to code the categorical features
occupation = tf.feature_column.categorical_column_with_hash_bucket(key='occupation', hash_bucket_size=5 )
occupation
```

```python
categorical_columns = [i for i in X.columns if X[i].dtype=='object']
print(categorical_columns)

categorical_features = [tf.feature_column.categorical_column_with_vocabulary_list(key=column,
                                                         vocabulary_list = df[column].unique())
                                                                                  for column in categorical_columns]
                        
```

```python
categorical_features[0]
```

** Create the continuous feature_columns for the continuous values using numeric_column **

```python
numeric_columns = [i for i in X.columns if X[i].dtype!='object']
print(numeric_columns)

numeric_features = [tf.feature_column.numeric_column(key=col) for col in numeric_columns]
```

```python
numeric_features
```

** Put all these variables into a single list with the variable name feat_cols **

```python
feat_cols = categorical_features + numeric_features
```

### Create Input Function

** Batch_size is up to you. But do make sure to shuffle!**

```python
input_func = tf.estimator.inputs.pandas_input_fn(x=X_train, y=y_train, batch_size=32, shuffle=True, num_epochs=1000)
```

#### Create your model with tf.estimator

**Create a LinearClassifier.(If you want to use a DNNClassifier, keep in mind you'll need to create embedded columns out of the cateogrical feature that use strings, check out the previous lecture on this for more info.)**

```python
checkpoint_path = "new_models/linear_model"
```

```python
linear_model = tf.estimator.LinearClassifier(feature_columns=feat_cols, n_classes=2, model_dir=checkpoint_path,
#                                              optimizer=tf.keras.optimizers.Ftrl(
#       learning_rate=0.1,
#       l1_regularization_strength=0.001
#     )
)
```

** Train your model on the data, for at least 5000 steps. **

```python
linear_model.train(input_fn=input_func, steps=5000)
```

### Evaluation

** Create a prediction input function. Remember to only supprt X_test data and keep shuffle=False. **

```python
pred_func = tf.estimator.inputs.pandas_input_fn(x=X_test, batch_size=len(X_test), shuffle=False, num_epochs=1)
```

** Use model.predict() and pass in your input function. This will produce a generator of predictions, which you can then transform into a list, with list() **

```python
predictions = list(linear_model.predict(input_fn=pred_func))
```

** Each item in your list will look like this: **

```python
predictions
```

** Create a list of only the class_ids key values from the prediction list of dictionaries, these are the predictions you will use to compare against the real y_test values. **

```python
final_prediction = [pred['class_ids'][0] for pred in predictions]
```

```python
final_probabilities = [pred['probabilities'][0] for pred in predictions]
```
